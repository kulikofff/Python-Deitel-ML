```python
# Task: how to train the model, if the columns in train and test data differs. It can happen:) 


from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
#Train dataset
dftrain = {'col1': [1, 2], 'col2': [3, 4], 'col3': [5, 6], 'y':[4,7]}
dftrain = pd.DataFrame(dftrain)
```

```python
dftrain
```

```python
y = dftrain['y']
```

```python
y
```

```python
dftrain = dftrain.drop(['y'], axis =1)
dftrain
```

```python
model = LinearRegression()
model.fit(dftrain, y)
```

```python
#Train dataset
dftest = {'col1': [1, 2], 'col2': [3, 4], 'col4': [11, 12], 'col55': [444,44]}
dftest = pd.DataFrame(dftest)
dftest
```

```python
res = model.predict(dftest)
res
```

```python
#Delete columns in bigger dataset that are not in smaller
dftrain_cols = dftrain.columns
dftest_cols = dftest.columns
dfval1 = dftest_cols.difference(dftrain_cols)
dfval1
dftest = dftest.drop(dfval1, axis=1)
dftest
```
